```python
y = file['Total Claim Amount']
y.head()

```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Train Linear Regression Model
from sklearn import linear_model

lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
```

```python
#predict and inspect results
results_for_test = lm.predict(X_test)

pd.concat([X,pd.Series(results_for_test)],axis=1).head()
```

```python
# Check the accuracy of the predictions
from sklearn.metrics import mean_squared_error, r2_score

# Check Mean Squared Error 
mse = mean_squared_error(y_test, results_for_test)
print(mse)
```

```python
# Check square root of Mean Squared Error
rmse = np.sqrt(mse)
print(rmse)
```

```python
# Check R squared score
r2 = r2_score(y_test, results_for_test)
r2
```
